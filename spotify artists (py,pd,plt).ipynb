```python
# import requests # Importa la biblioteca requests para realizar solicitudes HTTP.
import pandas as pd # Importa la biblioteca pandas para el análisis y manipulación de datos en formato de tabla.
import seaborn as sns # Importa la biblioteca seaborn para la visualización de datos estadísticos.
# import numpy as np # Importa la biblioteca numpy para realizar cálculos numéricos complejos.
import matplotlib.pyplot as plt # Importa la biblioteca matplotlib para crear gráficos y visualizaciones.
```

```python
#2023
html_2023 = pd.read_html('https://en.wikipedia.org/wiki/List_of_most-streamed_artists_on_Spotify')
df2023 = html_2023[3]
df_20artistas2023= df2023.iloc[:20]

#extraer nombres de artistas y almacenar en lista y eliminanar la última fila con un total general
ls_artistas2023 = df_20artistas2023[df_20artistas2023.columns[1]].values.tolist()
del ls_artistas2023[-1]
print(ls_artistas2023)

#mismo proceso pero con streaming 
ls_streams2023 = df_20artistas2023[df_20artistas2023.columns[2]].values.tolist()
del ls_streams2023[-1]

#converir datos de streaming en flotante y pasar a intenger
streams2023 = [int(float(s)) for s in ls_streams2023]
print(streams2023)
```

```python
#muestra un grafico de barras horizontales utilizando los datos previamente extraidos

y=streams2023[:20]
x=ls_artistas2023[:20]

#crear un grafico de barras horizontales
plt.barh(x, y)
plt.title('Top 10 artistas más escuchados 2023, en millones')

#colocar los valores sobre las barras correspondientes en el gráfico
for indice, valor in enumerate(y):
    plt.text(valor, indice,
             str(valor))
          
plt.show()
```

```python
#2022
html_2022 = pd.read_html('https://web.archive.org/web/20221017171921/https://en.wikipedia.org/wiki/List_of_most-streamed_artists_on_Spotify')
df2022 = html_2022[1]
df_20artistas2022 = df2022.iloc[:20]

ls_artistas2022 = df_20artistas2022[df_20artistas2022.columns[1]].values.tolist()
del ls_artistas2022[-1]
print(ls_artistas2022)

ls_streams2022 = df_20artistas2022[df_20artistas2022.columns[2]].values.tolist()
del ls_streams2022[-1]

streams2022 = [int(float(s)) for s in ls_streams2022]
print(streams2022)
```

```python
#2021
html_2021 = pd.read_html('https://web.archive.org/web/20210408013816/https://en.wikipedia.org/wiki/List_of_most-streamed_artists_on_Spotify')
df2021 = html_2021[2]
df_20artistas2021 = df2021.iloc[:20]

ls_artistas2021 = df_20artistas2021[df_20artistas2021.columns[1]].values.tolist()
del ls_artistas2021[-1]
print(ls_artistas2021)

ls_streams2021 = df_20artistas2021[df_20artistas2021.columns[2]].values.tolist()
del ls_streams2021[-1]

streams2021 = [int(float(s)) for s in ls_streams2021]
print(streams2021)
```

```python
#2020
html_2020 = pd.read_html('https://web.archive.org/web/20200428071326/https://en.wikipedia.org/wiki/List_of_most-streamed_artists_on_Spotify')
df2020 = html_2020[2]
df_20artistas2020 = df2020.iloc[:20]

ls_artistas2020 = df_20artistas2020[df_20artistas2020.columns[1]].values.tolist()
del ls_artistas2020[-1]
print(ls_artistas2020)

ls_streams2020 = df_20artistas2020[df_20artistas2020.columns[2]].values.tolist()
del ls_streams2020[-1]

streams2020 = [int(float(s)) for s in ls_streams2020]
print(streams2020)
```

```python
# Encontrar los artistas que aparecen en las cuatro listas
artistas_repetidos = []
for artista in ls_artistas2023:
    if artista in ls_artistas2022 and artista in ls_artistas2021 and artista in ls_artistas2020:
        artistas_repetidos.append(artista)

# Mostrar los artistas que aparecen en las cuatro listas
print("Artistas que aparecen en las cuatro listas:")
for artista in artistas_repetidos:
    print(artista)


# Pedir al usuario que ingrese un artista específico
while True:
    artista_buscado = input('Escribe el nombre del artista a buscar: ')
    
    if artista_buscado in artistas_repetidos:
        print('El artista', artista_buscado, 'se encuentra en las cuatro listas.')
        break
    else:
        print('El artista', artista_buscado, 'no se encuentra en las cuatro listas. Vuelve a intentarlo.')
        continue



s2023 = []
s2022 = []
s2021 = []
s2020 = []

#verificar si artista se encuentra en listas de cada agno,si se encuestra, almacenar la cantidad de streams correspondiente a cada agno 

if artista_buscado in ls_artistas2023:
    index = ls_artistas2023.index(artista_buscado)
    s2023.append(streams2023[index])
    
if artista_buscado in ls_artistas2022:
    index = ls_artistas2022.index(artista_buscado)
    s2022.append(streams2022[index])    
    
if artista_buscado in ls_artistas2021:
    index = ls_artistas2021.index(artista_buscado)
    s2021.append(streams2021[index])    
     
if artista_buscado in ls_artistas2020:
    index = ls_artistas2020.index(artista_buscado)
    s2020.append(streams2020[index])
      
print (s2020,s2021,s2022,s2023)



#representar los resultados en un diagrama de lineas


years = ['2020', '2021', '2022', '2023']
values = [s2020[0], s2021[0], s2022[0], s2023[0]]


    # Creamos la figura y los ejes
fig, ax = plt.subplots()

    # Creamos la línea del diagrama de líneas
ax.plot(years, values)

    # Creamos los puntos correspondientes a cada valor y su respectivo valor
for i in range(len(years)):
    ax.scatter(years[i], values[i], color='red')
    ax.annotate((values[i]), (years[i], values[i]), textcoords="offset points", xytext=(0,5), ha='center')

    # Añadimos etiquetas al diagrama
ax.set_xlabel('Año')
ax.set_ylabel('Streaming en millones')
ax.set_title(artista_buscado)


    # Mostramos el diagrama
plt.show()

#representar los resultados en un diagrama de barras verticales

    #funcion para agregar etiquetas con los valores correspondientes a cada barra
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]//2, y[i], ha = 'center')

    # Crear una nueva figura
fig = plt.figure()

    # Agregar un objeto de ejes a la figura
ax = fig.add_axes([0,0,1,1])

    # Definir las etiquetas para el eje x y los valores para el eje y
x=['Streams2020','Streams 2021','Streams 2022','Streams 2023']
y=[s2020[0],s2021[0],s2022[0],s2023[0]]


    # Crear el grafico de barras
barlist=plt.bar(x,y)

    # Establecer colores distintos en barras
barlist[0].set_color('r')
barlist[1].set_color('b')
barlist[2].set_color('g')
barlist[3].set_color('y')

    # Agregar etiquetas a las barras
addlabels(x,y)

    # Mostrar el grafico
plt.show()

```
